La funci√≥ **format_sky** us ajudar√† a visualitzar el tauler del problema.

````python
import numpy as np
from IPython.display import clear_output
import time

def format_sky(board, top, bottom, left, right):
    """
    Funci√≥ auxiliar per mostrar la matriu del Problema 1 en forma de graella.
    """
    _str = ""
    
    # Files
    for v in top:
        _str+="   "+ str(v)
    _str = "  " + _str + "\n"

    for idx, i in enumerate(board):
        _str+= "   +"+ ("-"*((board.shape[1]*4)-1)) +"+\n"
        _str+= f" {left[idx]} | "
        
        # Columnes
        for j in i:   
            if j!=0:
                _str+= str(j)+" | "
            else:
                _str+= "  | "
            
        _str+= f"{right[idx]} \n"
    _str+= "   +"+ ("-"*((board.shape[1]*4)-1))+"+\n"
    
    _bot = ""
    for v in bottom:
        _bot+="   "+str(v)
    _str += "  " + _bot
    
    return _str.replace('0',' ')
````

Veieu un exemple de com utilitzar aquesta funci√≥:


````python
# Amb aquesta funci√≥ es mostren totes les caselles del tauler
# Les caselles que contenen un zero es mostren buides

# Creem unes restriccions i un tauler 
top, bottom, left, right = [0,0,0], [3,0,0], [0,0,0], [0,0,1]
grid = np.zeros((len(left), len(top)), dtype='int')

print('Tauler:')
print(format_sky(grid, top, bottom, left, right))
````

La funci√≥ **delayed_print** us permet bloquejar temporalment un 'print' per si voleu veure les accions que fa a cada pas el vostre algorisme.<br>
No heu d'utilitzar-la a l'entrega ja que el programa tardar√† molt m√©s del necessari, utilitzeu-la nom√©s mentre programeu la soluci√≥.

````python
def delayed_print(grid, top, bottom, left, right, sleep_time=0.02):
    """
    Funci√≥ auxiliar que bloqueja temporalment la funci√≥ 'print' per veure l'evoluci√≥ de l'algorisme.
    """
    print(format_sky(grid, top, bottom, left, right))
    clear_output(wait=True)
    time.sleep(sleep_time)
````

````python
###
# LES VOSTRES FUNCIONS EXTRES AQU√ç
###

#Creo una funcio auxiliar que m'ajudar√† a comptar el nombre d'edificis visibles
def count_visible(llista):
    visibles = 0
    max_h = 0
    
    for h in llista:
        if h == 0:
            #si hi ha algun 0, encara no podem saber del cert quants edificis visibles hi ha 
            return None
        
        if h > max_h:
            max_h = h
            visibles += 1
            
    return visibles


# Amb els valors que hi ha fins ara (incloent zeros), √©s possible arribar a complir la restricci√≥?
# No √©s una validaci√≥ final: √©s un filtre que et diu si podria arribar a ser possible o no
def satisfies_petit(fila, restriccio):
    n = len(fila)
    
    # si la restriccio es 0 retornem true, pqe qualsevol combinacio ser√† real
    if restriccio == 0:
        return True

    #comptar√© els edificis visbles que tinc
    #no fem servir la funcio auxiliar comptadora d'edificis perqu√® aquella nomes compta si NO hi han zeros
    max_altura = 0
    edificis_visibles = 0
    ceros = 0
    
    for altura in fila:
        if altura == 0:
            ceros += 1
        else:
            if altura > max_altura:
                edificis_visibles += 1
                max_altura = altura

    # condicio 1: si veig mes edificis dels que puc veure
    if edificis_visibles > restriccio:
        return False
    
    # condicio 2: si no hi ha zeros ha de coincidir exactement
    if ceros == 0 and edificis_visibles != restriccio:
        return False
    
    # condicio 3: si tots els zeros fossin edificis mes alts no podriem arribar a la restriccios
    # cas [0, 4, 3, 1]
    # Encara que tots els zeros fossin edificis m√©s alts i visibles, no podr√≠em arribar a la restricci√≥
    if edificis_visibles + ceros < restriccio:
        return False
    
    #si no es cumpleix cap de les condicions, retornem un true final
    return True


def satisfies(grid, x, y, num, top, bottom, left, right):
    """
    Donat un tauler, un punt amb coordenades (x,y) i un nombre 'num', comprova si √©s possible assignar-lo a la posici√≥ (x,y) 
    donades les restriccions del problema.
    
    Params
    ------
    :grid: Una matriu de numpy
    :x,y: Un punt de la matriu on volem posar el nombre 'num'
    :num: Un nombre enter
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :b: Un bole√† True/False depenent de si el nombre 'num' pot ser col¬∑locat a la casella (x,y) o no.
    """
    
    #gid es la matriu de numpy, shape √©s una tupla amb les seves dimensions, per tant agafem el primer element de la tupla 
    #com que es tracta d'una matriu quadrada, num files = num columnes
    num_files = grid.shape[0]

    # Primer mirem si hi ha algun numero a la fila o columna que a estigui posat i sigui igual al num
    for i in range(num_files):
        #primer observem la seva columna
        if grid[i][y] == num:
            return False
    for j in range(num_files):
        #ara observem la seva fila
        if grid[x][j] == num:
            return False
        
    # Assignem el nombre a la casella temporalment, si no es possible que estigui en aquesta posicio, ja el trurem
    grid [x][y] = num
    
    
    # Ara comprovem si stisfa les restriccions de top o de bottom (es a dir, ens fixem amb la columna)
    columna = grid[:, y]
    
    if top[y] != 0 and not satisfies_petit(columna, top[y]):
        #Cas que NO tinguem la columna completa:
        grid[x][y] = 0
        return False
    if bottom[y] != 0 and not satisfies_petit(columna[::-1], bottom[y]):
        #Cas que NO tinguem la columna completa:
        grid[x][y] = 0
        return False
    
    
    # Ara comprovem si stisfa les restriccions de left o de right (es a dir, ens fixem amb la fila)
    fila = grid[x, :]
    
    if left[x] != 0 and not satisfies_petit(fila, left[x]):
        grid[x][y] = 0
        return False
    if right[x] != 0 and not satisfies_petit(fila[::-1], right[x]):
        grid[x][y] = 0
        return False
        
    #important tornar la posicio en el seu estat original (que es 0)
    grid[x][y] = 0
    
    #si ha arribat fins aqui, √©s que si que cumpleix totes les restriccions
    return True
    

def skyscrapper_backtracking(grid, top, bottom, left, right):  
    """
    Funci√≥ que implementa l'algorisme de backtracking
    
    Params
    ------
    :grid: Una matriu de tipus numpy
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :grid: La matriu plena amb la soluci√≥ correcta o b√© False en cas que no tingui soluci√≥.    
    """
    
    # NO esborreu aquestes dues l√≠nies, serveien per saber quants cops es crida aquesta funci√≥
    # ----------------------------------------------------------------------------------------
    global num_recursive_calls 
    num_recursive_calls += 1
    # ----------------------------------------------------------------------------------------
    
    # EL TEU CODI AQU√ç

    # si la matriu esta plena ha de retornar la matriu --> 
    # np.all(grid != 0) --> retorna si tots els espais de la matriu son diferent de 0 --> esta plena
    # si esta plena, √©s pqe l'hem pogut omplir, i si l'hem pogut omplir, √©s que t√© soluci√≥
    if np.all(grid != 0):
        return grid
    
    #prenem la llargada de les files/columnes
    n = grid.shape[0]
    
    for i in range(n):
        for j in range (n):
            #les caselles que ja estan omplertes ens les saltem, volem omplir les dels 0's
            if grid[i][j] == 0:
                # Probarem totes les al√ßades possibles
                for altura in range(1, n + 1):
                    if satisfies (grid, i, j, altura, top, bottom, left, right):
                        grid[i][j] = altura
                        
                        # Crida recursiva
                        result = skyscrapper_backtracking(grid, top, bottom, left, right)
                        if result is not False:
                            return result
                        
                        #part del backtarck: trec la solucio que no mha anat b√© (torno enrere) i en probar√© una altra
                        grid[i][j] = 0
                        
                #si arriba aqui, es pqe cap altura funciona, per tant retornem false        
                return False
            
    # si arribes aqui, √©s pqe no hi ha cap 0, la matriu est√† completa i √©s correcta
    return grid

def skyscrapper(top, bottom, left, right):
    """
    Funci√≥ principal del problema. Rep quatre llistes corresponents als nombres que hi ha fora del tauler.
    
    Params
    ------
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    """
    
    # Inicialitzem una matriu de zeros
    grid = np.zeros((len(left), len(top)), dtype='int')
    
    # Cridem a la funci√≥ que soluciona el problema mitjan√ßant backtracking
    sol = skyscrapper_backtracking(grid, top, bottom, left, right)   
    
    # Mostrem el resultat en el cas que trobem una soluci√≥ o mostrem un error en cas contrari.
    if sol is not False:
        print(format_sky(np.array(sol), top, bottom, left, right))
    else:
        print('No solution found')
# sincro   
````
