La funció **format_sky** us ajudarà a visualitzar el tauler del problema.

````python
import numpy as np
from IPython.display import clear_output
import time

def format_sky(board, top, bottom, left, right):
    """
    Funció auxiliar per mostrar la matriu del Problema 1 en forma de graella.
    """
    _str = ""
    
    # Files
    for v in top:
        _str+="   "+ str(v)
    _str = "  " + _str + "\n"

    for idx, i in enumerate(board):
        _str+= "   +"+ ("-"*((board.shape[1]*4)-1)) +"+\n"
        _str+= f" {left[idx]} | "
        
        # Columnes
        for j in i:   
            if j!=0:
                _str+= str(j)+" | "
            else:
                _str+= "  | "
            
        _str+= f"{right[idx]} \n"
    _str+= "   +"+ ("-"*((board.shape[1]*4)-1))+"+\n"
    
    _bot = ""
    for v in bottom:
        _bot+="   "+str(v)
    _str += "  " + _bot
    
    return _str.replace('0',' ')
````

Veieu un exemple de com utilitzar aquesta funció:


````python
# Amb aquesta funció es mostren totes les caselles del tauler
# Les caselles que contenen un zero es mostren buides

# Creem unes restriccions i un tauler 
top, bottom, left, right = [0,0,0], [3,0,0], [0,0,0], [0,0,1]
grid = np.zeros((len(left), len(top)), dtype='int')

print('Tauler:')
print(format_sky(grid, top, bottom, left, right))
````

La funció **delayed_print** us permet bloquejar temporalment un 'print' per si voleu veure les accions que fa a cada pas el vostre algorisme.<br>
No heu d'utilitzar-la a l'entrega ja que el programa tardarà molt més del necessari, utilitzeu-la només mentre programeu la solució.

````python
def delayed_print(grid, top, bottom, left, right, sleep_time=0.02):
    """
    Funció auxiliar que bloqueja temporalment la funció 'print' per veure l'evolució de l'algorisme.
    """
    print(format_sky(grid, top, bottom, left, right))
    clear_output(wait=True)
    time.sleep(sleep_time)
````

````python
def satisfies_petit_fakee(fila, restriccio):
    
    # si la restriccio es 0 retornem true
    if restriccio == 0:
        return True

    # casos extrems
    if restriccio == 1:
        if fila[0] != len(fila) + 1:
            return False
    if restriccio == len(fila) + 1:
        for i in range (len(fila)):
            if fila[i] != i+1:
                return False


    edificis_visibles = 0
    zeros = 0
    visible = fila[0]
    if visible != 0:
        edificis_visibles += 1

    for i in range(1, len(fila)):
        if (visible < fila[i]):
            visible = fila[i]
            if visible != 0:
                edificis_visibles += 1
        if fila[i] == 0:
            zeros += 1

    if edificis_visibles > restriccio:
        return False
    if zeros == 0 and edificis_visibles != restriccio:
        return False
    
    return True
````

````python
###
# LES VOSTRES FUNCIONS EXTRES AQUÍ
###

#Creo una funcio auxiliar que m'ajudarà a comptar el nombre d'edificis visibles
def count_visible(llista):
    visibles = 0
    max_h = 0
    
    for h in llista:
        if h == 0:
            #si hi ha algun 0, encara no podem saber del cert quants edificis visibles hi ha 
            return None
        
        if h > max_h:
            max_h = h
            visibles += 1
            
    return visibles


# Amb els valors que hi ha fins ara (incloent zeros), és possible arribar a complir la restricció?
# No és una validació final: és un filtre que et diu si podria arribar a ser possible o no
def satisfies_petit(fila, restriccio):
    n = len(fila)
    
    # si la restriccio es 0 retornem true, pqe qualsevol combinacio serà real
    if restriccio == 0:
        return True

    #comptaré els edificis visbles que tinc
    #no fem servir la funcio auxiliar comptadora d'edificis perquè aquella nomes compta si NO hi han zeros
    max_altura = 0
    edificis_visibles = 0
    ceros = 0
    
    for altura in fila:
        if altura == 0:
            ceros += 1
        else:
            if altura > max_altura:
                edificis_visibles += 1
                max_altura = altura

    # condicio 1: si veig mes edificis dels que puc veure
    if edificis_visibles > restriccio:
        return False
    
    # condicio 2: si no hi ha zeros ha de coincidir exactement
    if ceros == 0 and edificis_visibles != restriccio:
        return False
    
    # condicio 3: si tots els zeros fossin edificis mes alts no podriem arribar a la restriccios
    # cas [0, 4, 3, 1]
    # Encara que tots els zeros fossin edificis més alts i visibles, no podríem arribar a la restricció
    if edificis_visibles + ceros < restriccio:
        return False
    
    #si no es cumpleix cap de les condicions, retornem un true final
    return True


def satisfies(grid, x, y, num, top, bottom, left, right):
    """
    Donat un tauler, un punt amb coordenades (x,y) i un nombre 'num', comprova si és possible assignar-lo a la posició (x,y) 
    donades les restriccions del problema.
    
    Params
    ------
    :grid: Una matriu de numpy
    :x,y: Un punt de la matriu on volem posar el nombre 'num'
    :num: Un nombre enter
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :b: Un boleà True/False depenent de si el nombre 'num' pot ser col·locat a la casella (x,y) o no.
    """
    
    #gid es la matriu de numpy, shape és una tupla amb les seves dimensions, per tant agafem el primer element de la tupla 
    #com que es tracta d'una matriu quadrada, num files = num columnes
    num_files = grid.shape[0]
    

    # hem de comprovar que les posicions son menors


    # Primer mirem si hi ha algun numero a la fila o columna que a estigui posat i sigui igual al num
    for i in range(num_files):
        #primer observem la seva columna
        if grid[i][y] == num:
            return False
    for j in range(num_files):
        #ara observem la seva fila
        if grid[x][j] == num:
            return False
        
    # Assignem el nombre a la casella temporalment, si no es possible que estigui en aquesta posicio, ja el trurem
    grid [x][y] = num
    
    
    # Ara comprovem si stisfa les restriccions de top o de bottom (es a dir, ens fixem amb la columna)
    columna = grid[:, y]
    
    if top[y] != 0:
        #Cas que NO tinguem la columna completa:
        if not satisfies_petit(columna, top[y]):
            grid[x][y] = 0
            return False
        
        #Cas que tinguem la columna completa:
        visibles = count_visible(columna)
        #si tenim tota la fila plena, pero al provar de posar un nombre no surt be, hi posem un 0 alla
        if visibles is not None and visibles != top[y]:
            grid[x][y] = 0
            return False
    
        
    if bottom[y] != 0:
        #Cas que NO tinguem la columna completa:
        if not satisfies_petit(columna[::-1], bottom[y]):
            grid[x][y] = 0
            return False
        
        #Cas que tinguem la columna completa:
        visibles = count_visible(columna[::-1])
        if visibles is not None and visibles != bottom[y]:
            grid[x][y] = 0
            return False
    
    
    # Ara comprovem si stisfa les restriccions de left o de right (es a dir, ens fixem amb la fila)
    fila = grid[x, :]
    
    if left[x] != 0:
        #Cas que NO tinguem la fila completa:
        if not satisfies_petit(fila, left[x]):
            grid[x][y] = 0
            return False
        
        #Cas que tinguem la fila completa:
        visibles = count_visible(fila)
        if visibles is not None and visibles != left[x]:
            grid[x][y] = 0
            return False

    if right[x] != 0:
        #Cas que NO tinguem la fila completa:
        if not satisfies_petit(fila[::-1], right[x]):
            grid[x][y] = 0
            return False
        
        #Cas que tinguem la fila completa:
        visibles = count_visible(fila[::-1])
        if visibles is not None and visibles != right[x]:
            grid[x][y] = 0
            return False

    #important tornar la posicio en el seu estat original (que es 0)
    grid[x][y] = 0
    
    #si ha arribat fins aqui, és que si que cumpleix totes les restriccions
    return True
    

def skyscrapper_backtracking(grid, top, bottom, left, right):  
    """
    Funció que implementa l'algorisme de backtracking
    
    Params
    ------
    :grid: Una matriu de tipus numpy
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :grid: La matriu plena amb la solució correcta o bé False en cas que no tingui solució.    
    """
    
    # NO esborreu aquestes dues línies, serveien per saber quants cops es crida aquesta funció
    # ----------------------------------------------------------------------------------------
    global num_recursive_calls 
    num_recursive_calls += 1
    # ----------------------------------------------------------------------------------------
    
    # EL TEU CODI AQUÍ

    # si la matriu esta plena ha de retornar la matriu --> 
    # np.all(grid != 0) --> retorna si tots els espais de la matriu son diferent de 0 --> esta plena
    # si esta plena, és pqe l'hem pogut omplir, i si l'hem pogut omplir, és que té solució
    if np.all(grid != 0):
        return grid
    
    #prenem la llargada de les files/columnes
    n = grid.shape[0]
    
    for i in range(n):
        for j in range (n):
            #les caselles que ja estan omplertes ens les saltem, volem omplir les dels 0's
            if grid[i][j] == 0:
                # Probarem totes les alçades possibles
                for altura in range(1, n + 1):
                    if satisfies (grid, i, j, altura, top, bottom, left, right):
                        grid[i][j] = altura
                        
                        # Crida recursiva
                        result = skyscrapper_backtracking(grid, top, bottom, left, right)
                        if result is not False:
                            return result
                        
                        #part del backtarck: trec la solucio que no mha anat bé (torno enrere) i en probaré una altra
                        grid[i][j] = 0
                        
                #si arriba aqui, es pqe cap altura funciona, per tant retornem false        
                return False
            
    # si arribes aqui, és pqe no hi ha cap 0, la matriu està completa i és correcta
    return grid

def skyscrapper(top, bottom, left, right):
    """
    Funció principal del problema. Rep quatre llistes corresponents als nombres que hi ha fora del tauler.
    
    Params
    ------
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    """
    
    # Inicialitzem una matriu de zeros
    grid = np.zeros((len(left), len(top)), dtype='int')
    
    # Cridem a la funció que soluciona el problema mitjançant backtracking
    sol = skyscrapper_backtracking(grid, top, bottom, left, right)   
    
    # Mostrem el resultat en el cas que trobem una solució o mostrem un error en cas contrari.
    if sol is not False:
        print(format_sky(np.array(sol), top, bottom, left, right))
    else:
        print('No solution found')
# sincro   
````

````python
num_recursive_calls = 0
t0 = time.perf_counter()
skyscrapper([0,0,0],
            [3,0,0],
            [0,0,0],
            [0,0,1])
t1 = time.perf_counter()
print('Crides recursives:', num_recursive_calls)
print(f'Temps: {t1-t0:.2f}s')
````
